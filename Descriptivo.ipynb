```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

```python
r="winequality-red.csv"
datos=pd.read_csv(r,sep=";")

#Variables predictoras
X=datos.drop("quality",axis=1)
X.describe()

X.alcohol.hist(color="#2ecc98")
plt.xlabel("Volumen de alcohol")
plt.ylabel("Frecuencia")
plt.title("Volumen de alcoho-frecuencia")
plt.show()
```

```python
Des=X.describe()
Des.to_csv("Descriptivo.csv",sep=";",decimal=",")
sts.median(X["alcohol"])
```

```python
y=datos["quality"]
sns.countplot(y,palette="hls")
plt.title("Calidad de los vino-frecuencia")
plt.xlabel("Calidad")
plt.ylabel("Frecuencia")
plt.title("Frecuencia en la calidad de vinos")
plt.savefig("Frecuencias vinos")
plt.show()
```
